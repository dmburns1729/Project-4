## Load Data

```python
import glob
q = "Data/Chicago/Chicago-Crime_*.csv"
chunked_files = sorted(glob.glob(q))
# Showing the first 5 
chunked_files[:5]
```

```python
## Loading all files as df and appending to a list
df_list = []
for file in chunked_files:
    temp_df = pd.read_csv(file, index_col=0)
    df_list.append(temp_df)
    
## Concatenating the list of dfs into 1 combined
df_combined = pd.concat(df_list)
df_combined
```

```python
df_ts.groupby("Primary Type").size().head()
```

```python
# create new column for each crime
## making our dictionary 
CRIMES = {}
crime = 'THEFT'
crime
```

```python
# I. Save a temp df of just the rows that match the crime
temp = df_ts.loc[ df_ts['Primary Type']==crime]
temp.head(3)
```

```python
# II. Resample the temp DataFrame as Daily data (crime counts)
# and keep ONLY the .size()
temp_res = temp.resample("D").size()
temp_res
```

```python
#III. Save the temporary DataFrame in the dictionary, #using the crime description as the key.
CRIMES[crime] = temp_res.copy()
CRIMES.keys()
```

```python
CRIMES['THEFT']
```

```python
# 1. get list of  unique crime descriptions
crime_list = df_ts['Primary Type'].unique()
crime_list
```

```python
# 2. Create an empty dictionary  
CRIMES = {}
# 3. Loop through the list of crimes
for crime in crime_list:
    # I. Save a temp df of just the rows that match the crime
   temp = df_ts.loc[ df_ts['Primary Type']==crime].copy()
    # II. Resample the temp DataFrame as Daily data (crime counts) 
    # and keep ONLY the .size() 
   temp_res = temp.resample("D").size() 
   #III. Save the temporary DataFrame in the dictionary,
   #using the crime description as the key.
   CRIMES[crime] = temp_res.copy() 
CRIMES.keys()
```

```python
CRIMES['RITUALISM']
```

```python
crimes_df = pd.DataFrame(CRIMES)
crimes_df
```

```python
crimes_df.isna().sum()
```

```python
## fill null values with 0
crimes_df = crimes_df.fillna(0)
crimes_df
```

```python
crimes_df.plot(grid=True,alpha=0.7);
```

```python
crimes_df.plot(subplots=True,figsize=(10,60),grid=True);
```

```python
# Resample as yearly data

crimes_annual = crimes_df.resample("Y").sum()
crimes_annual
```

```python
crimes_annual.plot(subplots=True,figsize=(10,60),grid=True);
```

### Answer

Overall crime is decreasing.  However, there are some crimes that are increasing while the general trend is down.  They are: weapons violation, deceptive practice, interference with public officer, homicide, stalking, criminal sexual assault, obscenity, public indecency, concealed carry  license violation, non-criminal, and human trafficking.  Some of these have very few occurrences so they do not change the overall trend. 

## 4) Comparing Months:
- What months have the most crime? What months have the least?
- Are there any individual crimes that do not follow this pattern? If so, which crimes?

```python
## Extracting the Month
df_ts['Month'] = df_ts.index.month
df_ts['MonthName'] = df_ts.index.month_name()
df_ts.head(3)
```

```python
# saving lookup for number vs name
unique_months = df_ts.drop_duplicates(subset=['Month']).sort_values('Month')
month_lookup =dict(zip(unique_months['Month'],unique_months['MonthName']))
month_lookup
```

```python
sns.countplot(data=df_ts, x='MonthName',palette='dark',order=month_lookup.values());

```

### Answer - part 1

February has the least crime and July and August have the most

```python
for crime in crime_list:
    print(f"{crime}")
    crime_type = df_ts[df_ts['Primary Type'] == crime]
    sns.countplot(data=crime_type, x='MonthName',palette='dark',order=month_lookup.values());
    plt.show()
```

### Answer - part 2

Crime generally follows the overall distribution with a couple of exceptions:

Other Offense is highest in January and generally decreases throughout the year

Battery is highest in May through July

Assault is highest in May through July

Gambling is low in the winter and high in the summer peaking in August

Offense involving children is highest in January

DECEPTIVE PRACTICE is highest in January

NARCOTICS generally declines as the year progresses

KIDNAPPING is highest in May and June

PROSTITUTION is highest in January

The crimes with very few data points are all over the map.
